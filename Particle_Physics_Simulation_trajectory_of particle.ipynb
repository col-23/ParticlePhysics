```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
q = 1.6e-19  # Charge of the particle (Coulombs)
charge_sign = 1 if q > 0 else -1
m = 9.11e-31  # Mass of the particle (kg)
B_x, B_y, B_z = 0.4, 0.4, 0.4  # Magnetic field components (Tesla)
time_step = 8e-11  # Time step (seconds)
total_time = 1e-8  # Total simulation time (seconds)
Particle = "Positron"

# Initial conditions
v_x, v_y, v_z = 1e6,  1e6, 1e6  # Initial velocity components (m/s)

# Arrays to store the trajectory
times = np.arange(0, total_time, time_step)
x_positions, y_positions, z_positions = np.array([]), np.array([]), np.array([])

# Magnetic field vector
B = np.array([B_x, B_y, B_z])

v_perpendicular = np.sqrt(v_x**2 + v_y**2)
# Calculate the radius using the formula
R = (m * v_perpendicular) / (np.absolute(q)* B_z)

for t in times:
    # Update position using the circular trajectory formula
    x = R * np.cos(charge_sign*2 * np.pi * (t / ((2 * np.pi * R)/v_perpendicular)))
    y = R * np.sin(charge_sign*2 * np.pi * (t / ((2 * np.pi * R)/v_perpendicular)))
    z = v_z * t

    # Store positions
    x_positions = np.append(x_positions, x)
    y_positions = np.append(y_positions, y)
    z_positions = np.append(z_positions, z)
# Plot the 3D trajectory
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = "3d")
ax.plot(x_positions, y_positions, z_positions, label= Particle+" Trajectory in B-field")

# Add labels and title
ax.set_title("Trajectory of a "+Particle+" in a Magnetic Field")
ax.set_xlabel("x-position (m)")
ax.set_ylabel("y-position (m)")
ax.set_zlabel("z-position (m)")

ax.grid(True)
ax.legend()

# Show the plot
plt.show()
```
